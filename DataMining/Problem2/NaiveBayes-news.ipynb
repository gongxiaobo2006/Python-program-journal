```python
import os 
from sklearn.naive_bayes import GaussianNB
from sklearn.feature_extraction.text import CountVectorizer
import glob
import numpy as np

fcf = glob.glob('/Users/Oliver/Desktop/香港大學研究生/Data Mining/Assignment2/dataset-news/*')
fcf = sorted(fcf)

message=[]
for i in range(len(fcf)):
    text = open(fcf[i],'r')
    raw_message = text.read()
    raw_message = raw_message.rstrip('\n')
    message=message + [raw_message]

count_vect = CountVectorizer(stop_words={'a','the','your','my','her','his','you','I','me','he','she','it','is','are','and','they','we','if','of'})
#count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(message)
X_train_counts.shape
X = X_train_counts.toarray()
print X
```

```python
path = '/Users/Oliver/Desktop/香港大學研究生/Data Mining/Assignment2/dataset-news/' 
dir = os.listdir(path)              # dir is directory
file_name=[]
for d in dir:                        # d is file name 
    file_name=file_name+[d]
file_name = sorted(file_name)
file_seq = []
for d in file_name:
    if 'auto' in d:
        file_seq.append(0)
    elif 'baseball' in d:
        file_seq.append(1)
    elif 'electronics' in d:
        file_seq.append(2)
    elif 'hockey' in d:
        file_seq.append(3)
    elif 'ibm-hw' in d:
        file_seq.append(4)
    elif 'mac-hw' in d:
        file_seq.append(5)
    elif 'moto' in d:
        file_seq.append(6)
    elif 'pol-gun' in d:
        file_seq.append(7)
print file_seq
```

```python
from sklearn.naive_bayes import GaussianNB
def gausNB(x_train,y_train,x_test,y_test):
    gnb = GaussianNB()
    gnb.fit(x_train, y_train)
    y_pred = gnb.predict(x_test)
    print "Number of mislabeled points out of a total %d points : %d" % \
    (x_test.shape[0],(y_test != y_pred).sum())
    accuracy=(y_test != y_pred).sum()
    return accuracy

from sklearn.naive_bayes import MultinomialNB
def MultiNB(x_train,y_train,x_test,y_test):
    mnb = MultinomialNB(alpha=1.0) 
    mnb.fit (x_train, y_train) 
    y_pred = mnb.predict(x_test)
    print "Number of mislabeled points out of a total %d points : %d" % \
    (x_test.shape[0],(y_test != y_pred).sum())
    accuracy=(y_test != y_pred).sum()
    return accuracy
```

```python
import numpy as np
from sklearn.model_selection import KFold

X = X
y = file_seq
kf = KFold(n_splits=10,shuffle=True)
counter = 1
accuracy=0
for train, test in kf.split(X):
    print "%d Fold" % counter
    x_train, x_test = np.array(X)[train], np.array(X)[test]
    y_train, y_test = np.array(y)[train], np.array(y)[test]
    accuracy_gausNB = gausNB(x_train,y_train,x_test,y_test)+accuracy
    accuracy_MultiNB = MultiNB(x_train,y_train,x_test,y_test)+accuracy
    counter += 1
    # train and predict using your model
    # count the accuracy
accuracy_gausNB = float(accuracy_gausNB)
accuracy_gausNB = 1-(accuracy_gausNB/800)
accuracy_MultiNB = float(accuracy_MultiNB)
accuracy_MultiNB = 1-(accuracy_MultiNB/800)
print "accuracy of Gaussian naive bayes is %f" % (accuracy_gausNB)
print "accuracy of Multinomial naive bayes is %f" % (accuracy_MultiNB)
# print the average accuracy
```
